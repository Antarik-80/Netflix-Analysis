### Q2. In which year highest number of TV shows and Movies were released

```python
df.dtypes
```

```python
df['Date_N'] = pd.to_datetime(df['Release_Date'], errors='coerce')
```

```python
df[pd.to_datetime(df['Release_Date'], errors='coerce').isna()]
```

```python
df = df.dropna(subset=['Date_N'])
```

```python
df
```

```python
df.shape
```

```python
df['Date_N'].dt.year.value_counts()
```

```python
df['Date_N'].dt.year.value_counts().plot(kind = "bar" , ylabel = 'Revenue', title = 'revenue vs date' )
```
